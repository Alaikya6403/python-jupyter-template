### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.

import matplotlib.pyplot as plt

# Create a 2x2 subplot layout
fig, axes = plt.subplots(2, 2, figsize=(8, 8))

# Sample data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
sizes = [40, 25, 20, 15]
labels = ['Python', 'Java', 'C++', 'JavaScript']

# First subplot: Line plot
axes[0, 0].plot(x, y, color='blue')
axes[0, 0].set_title('Line Plot')

# Second subplot: Scatter plot
axes[0, 1].scatter(x, y, color='red')
axes[0, 1].set_title('Scatter Plot')

# Third subplot: Bar plot
axes[1, 0].bar(labels, sizes, color='green')
axes[1, 0].set_title('Bar Plot')

# Fourth subplot: Pie chart
axes[1, 1].pie(sizes, labels=labels, autopct='%1.1f%%', colors=['blue', 'green', 'red', 'orange'], startangle=140)
axes[1, 1].set_title('Pie Chart')

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```
